```python
from datasets import load_dataset

# Load the HotpotQA dataset
dataset = load_dataset("hotpot_qa", "distractor")

# Check the available splits
print("Available splits:", dataset.keys())

# Get the training split
train_data = dataset["train"]

# Display a sample from the training data
sample = train_data[0]
print("\nSample Data:")
print(f"Question: {sample['question']}")
print(f"Context: {sample['context']}")
print(f"Answer: {sample['answer']}")
print(f"Supporting Facts: {sample['supporting_facts']}")

# Optional: iterate through a few samples
print("\nIterating through some samples:")
for i in range(5):  # adjust the range for more samples
    data = train_data[i]
    print(f"Sample {i+1}")
    print("Question:", data["question"])
    print("Answer:", data["answer"])
    print()
```

```python

for k,v in sample.items():
    if (k=="context"):
        print(k+":")
        for i in range(len(v["title"])):
            a = v["title"][i] 
            b = "\n".join([tmp.strip() for tmp in v["sentences"][i]])
            print(f"{a}: \n{b}")
            print("\n")
    else:
        print(f"{k}: {v}")
```
